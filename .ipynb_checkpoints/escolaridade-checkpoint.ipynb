# Limpeza e Preparação de Dados

Removemos dados ausentes e os tipos de cada coluna, poís será útil para análise.

```python
def stats_dados_faltantes(df: pd.DataFrame) -> None:

  stats_dados_faltantes = []
  for col in df.columns:
    if df[col].isna().any():
      qtd, _ = df[df[col].isna()].shape
      total, _ = df.shape
      dict_dados_faltantes = {col: {'quantidade': qtd, "porcentagem": round(100 * qtd/total, 2)}}
      stats_dados_faltantes.append(dict_dados_faltantes)

  for stat in stats_dados_faltantes:
    print(stat)
```

```python
stats_dados_faltantes(df=df)
```

```python
df.dropna(inplace=True)
```

# Distribuição de escolaridade dos clientes

A análise da distribuição de escolaridade dos clientes do banco revelou informações interessantes sobre o perfil educacional dos usuários. Observamos uma diversidade significativa nos níveis de escolaridade.

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
plt.style.use('dark_background')

dados = {'escolaridade': ['sem educacao formal', 'ensino medio', 'graduacao', 'mestrado', 'doutorado']}
contagem_escolaridade = df['escolaridade'].value_counts()
cores = ['#3b0503', '#f67704', '#f85313', '#fedc57', '#9ecfbc']


contagem_escolaridades = df['escolaridade'].value_counts()
plt.figure(figsize=(7, 7))
plt.pie(contagem_escolaridade, startangle=90, autopct='%1.1f%%', colors=cores, wedgeprops={'edgecolor': 'black', 'linewidth': 3.0} )
plt.title('Distribuição de escolaridade dos clientes')
plt.legend(title="Escolaridades", labels=contagem_escolaridades.index, bbox_to_anchor=(1, 0), loc="lower left")


plt.show()
```

Percebe-se que a maioria dos clientes já tem mestrado em sua escolaridade e, logo em seguida, clientes que têm o ensino médio completo. 
Somando esses dois grupos, equivale a metade da distribuição da escolaridade.

# Escolaridade dos clientes Adimplentes

Nessa análise, iremos ver a distribuição de escolaridade dos clientes que está em dia com suas obrigações fiscais.

```python
df_adimplentes = df[df['default'] == 0]
contagem_adimplentes = df_adimplentes['escolaridade'].value_counts()
cores_adimplentes = ['#2f2e30', '#e84b2c', '#e6d839', '#7cd164', '#2eb8ac']


plt.style.use('dark_background')
plt.figure(figsize=(8, 8))
plt.pie(contagem_adimplentes, startangle=90, autopct='%1.1f%%', colors=cores_adimplentes, wedgeprops={'edgecolor': 'black', 'linewidth': 3.0})
plt.title('Distribuição dos Clientes das Escolaridade dos Adimplentes')
plt.legend(title="Legenda", labels=contagem_escolaridades.index, bbox_to_anchor=(1, 0), loc="lower left")

plt.show()
```

Em relação a distribuição geral, as únicas pequenas diferenças é o percentual dos clientes com doutorado, com graduação e os de sem educação formal.

# Escolaridade dos clientes inadimplentes

Os clientes onde não cumpre com suas obrigações financeiras dentro dos prazos estabelecidos tem sua análise.

```python
df_inadimplentes = df[df['default'] == 1]
contagem_inadimplentes = df_inadimplentes['escolaridade'].value_counts()
cores_inadimplentes = ['#4e031e', '#5d2d4e', '#5a4c6e', '#447390', '#05a1ad']


plt.style.use('dark_background')
plt.figure(figsize=(8, 8))
plt.pie(contagem_inadimplentes, startangle=90, autopct='%1.1f%%', colors=cores_inadimplentes, wedgeprops={'edgecolor': 'black', 'linewidth': 3.0})
plt.title('Distribuição dos Clientes das Escolaridade dos Inadimplentes')
plt.legend(title="Legenda", labels=contagem_escolaridades.index, bbox_to_anchor=(1, 0), loc="lower left")

plt.show()
```
